This notebook is used to load extremely large files (> 500 MB) directly from the S drive of the SAFE Desktop to the crunchr analytics platform.  
The loaded file is then saved into the crunchr data user volume specified by directory.

Author: Vina Ro

```python
# Import packages to load file
import getpass
from SciServer import Authentication
import smbclient
import pandas as pd
from pathlib import Path
```

```python
def load_file(load_filename, save_filename, load_dir, save_dir):
    '''
    Input params:
    - filename: target filename (csv)
    - load_dir: the directory of the target file (S Drive)
    - save_dir: filepath of where you want to save the file
    '''
    # Setup JHED login credentials
    myUserName = Authentication.getKeystoneUserWithToken(Authentication.getToken()).userName
    passwd = getpass.getpass('Password for ' + myUserName + ': ')
    user = "win\\" + myUserName
    usernamePwd = user+"%"+passwd
    
    # Connect to S drive using smbclient
    smb = smbclient.SambaClient(server="cloud.nas.jh.edu", share="sddesktop$",
                                username=usernamePwd, password=passwd, domain='win')
    print('Downloading file ------')
    smb.download(load_dir + '/' + load_filename, load_filename)
    # df = pd.read_csv(filename)
    print('Reading file ------')
    df = pd.read_csv(load_filename, compression = 'gzip') # If file is compressed
    print('Saving file ------')
    df.to_csv(save_dir / save_filename)
    return df
```

```python
# Set path
file_path = Path('PMAP_Final_Cohort_Vina.ipynb').resolve()
data_path = file_path.parent.parent.parent.parent.parent.joinpath('Data/jbergma8/IRB_271579_Faraday/IRB-271579-v3-DEID-230117-update')

# Main
load_file("labs.csv.gz", 'labs.csv', "/Faraday_Surg_Outcome/data/IRB-271579-v3-DEID-230117-update", data_path)
```
